```python
#Initiate the lrate
lrate = 0.001
```

#### Equation of derivative of loss with respect to weight
<br>
$mse = {1\,/\,n}\sum_{i=1}^n ({y\_actual}_{i}-(wx_{i}+b))^{2} $

```python
#loss fucntion calculate the mean square error
def loss_fn (y, y_predicted):
    n = len(y)
    s = 0
    for i in range(len(y)):
        s += (y[i] - y_predicted[i])**2
    return (1/n)*s
```

#### Equation of derivative of loss with respect to weight
<br>
$\frac{dl}{dw}= {2\,/\,n}\sum_{i=1}^n -x_{i}*(y_{i}-\tilde{y}_{i}) $


```python
#Derivative of loss w.r.t weight
def dldw(x, y, y_predicted):
    n = len(y)
    s = 0
    for i in range(len(y)):
        s += -x[i]*(y[i] - y_predicted[i])
    return (2/n)*s
```

#### Equation of derivative of loss with respect to intercept
<br>
$\frac{dl}{db}= {2\,/\,n}\sum_{i=1}^n -(y_{i}-\tilde{y}_{i}) $

```python
#Derivative of loss w.r.t intercept
def dldb(y, y_predicted):
    n = len(y)
    s = 0
    for i in range(len(y)):
        s += -(y[i] - y_predicted[i])
    return (2/n) * s
```

#### Equation of predicted y
<br>
$\tilde{y}  = weight * x + intercept$
<br><br>
where 
<br>
$weight$ is the rate of change and,
<br>
$x$ is the value of x and,
<br>
$intercept$ is the y intercept


```python
#Calculate the value of predicted y
def predicted_y(weight,x,intercept):
    y_lst = []
    for i in range(len(x)):
        y_lst.append(weight@x[i] + intercept) #dot product btween weight and input
    #return y_lst
    return np.array(y_lst)
```

## Model Training

##### Update the weight and intercept till we reach the global minima

$w_{new} = w_{old} - learning\,rate * \sum_{i=1}^n {dl\,/\,dw}$,
<br><br>
$b_{new} = b_{old} - learning\,rate * \sum_{i=1}^n {dl\,/\,db}$,
<br><br>
where 
<br>
${dl\,/\,dw}$ is deriative of loss w.r.t weight,
<br>
${dl\,/\,db}$ is deriative of loss w.r.t intercept,
<br>
$n$ is the total number of records

```python
def train_model(x, y, lr, epoch):
    loss_list = []
    weight_vector = [0, 0, 0, 0, 0]
    intercept = 0
    
    for i in range(epoch):
        y_predicted = predicted_y(weight_vector, x, intercept)
        weight_vector = weight_vector - lrate *dldw(x, y, y_predicted)  #update weight
        intercept = intercept - lrate * dldb(y, y_predicted)    #update bias
        loss_function = loss_fn(y, y_predicted)
        loss_list.append(loss_function)
        print("Epoch: " + str(i + 1) + "/" + str(epoch))
        print("Weight: ", weight_vector)
        print("Intercept: " + str(intercept))
        print("Loss: " + str(loss_function) + "\n")
    
    return weight_vector, intercept, loss_list
```

```python
#Number of epochs
epoch = 2000

#Train the model
model = train_model(np.array(x_train), np.array(y_train), lrate, epoch)
```

```python
#Print the weight 
print("weight: ", model[0])
#Print the intercept
print("intercept: ", model[1])
```

```python
#Plot the training loss against epoch graph
epoch_range = np.linspace(0, 2000, 2000)

plt.figure(figsize = (15, 10))
plt.minorticks_on()
plt.plot(epoch_range, model[2], '-')
plt.grid(b = True, which = 'both')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.minorticks_on()
plt.title('Training Loss against Epoch', fontsize = 20, fontweight ='bold')
plt.ylabel('Training Loss', fontsize = 15, fontweight ='bold')
plt.xlabel('Epoch', fontsize = 15, fontweight ='bold')
plt.show()
```

# Model Testing

```python
#Predict the y values of x_test
y_pred = predicted_y(model[0], np.array(x_test), model[1])
y_pred
```
